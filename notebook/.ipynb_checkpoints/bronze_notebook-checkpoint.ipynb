```python
from pyspark.sql import SparkSession
import os
import json
import requests
import time
from pyspark.sql.functions import current_timestamp, lit

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("blue-owls-assessment") \
    .getOrCreate()

spark.version  # confirm Spark is running

API_BASE_URL = os.environ.get("API_BASE_URL", "http://mock-api:8000")
API_USERNAME = os.environ.get("API_USERNAME", "candidate")
API_PASSWORD = os.environ.get("API_PASSWORD", "blue-owls-2026")
MANIFEST_PATH = "blue-owls-de-assessment/submission/ingestion.json"
```

```python
# Function to get the tokens
def get_token():
    resp = requests.post(f"{API_BASE_URL}/api/v1/auth/token", json={
    "username": API_USERNAME,
    "password": API_PASSWORD
    })
    token = resp.json()["access_token"]
    return token
    
```

```python
# Function to get and update manifest file to track the proceed pages
def read_manifest():
    if not os.path.exists(MANIFEST_PATH):
        return {}
    with open(MANIFEST_PATH, "r") as f:
        return json.load(f)
        

def update_manifest(manifest):
    with open(MANIFEST_PATH, "w") as f:
        json.dump(manifest, f, indent=2)
```

```python
#Get data from the endpoint
def get_data(url,token,endpoint):
    headers = {"Authorization": f"Bearer {token}"}
    max_retries=3
    failed_page=[]
    all_data=[]
    page=1
    completed_pages=set()
    manifest = read_manifest()
    endpoint_state = manifest.get(endpoint, {"completed_pages": []})
    completed_pages = set(endpoint_state["completed_pages"])
    
    
    while True:
        retries=0
        if page in completed_pages:
            page+=1
            continue
        try:
                while retries<max_retries  :
                    response=requests.get(url, headers=headers, params={"page": page, "page_size": 1000}).json()
                    if "data" in response:
                        break
                    # print(f"Retry {retries+1} for page {page}")
                    retries+=1
                    time.sleep(1)
                if "data" not in response:
                    print(f"Skipping page {page} after retries:",response)
                    page+=1
                    failed_page.append(page)
                    continue
        
                if page>=response["pagination"]["total_pages"]:
                    break
                all_data.extend(response.get("data",[]))
                completed_pages.add(page)
                page+=1
        except Exception as e:
            print(f"{endpoint} page {page} failed: {e}")
            break   # stop this endpoint only
    manifest[endpoint] = {
                "completed_pages": list(completed_pages)
            }
    print(completed_pages)
    update_manifest(manifest)  

    return all_data
        

        
```

```python
# Ingesting Orders Data to Bronze Layer
token=get_token()
alldata=get_data(f"{API_BASE_URL}/api/v1/data/orders",token,"orders")
if alldata:
    df=spark.createDataFrame(alldata)
    print(df.count())
    df=df.withColumn("_ingested_at", current_timestamp()) \
         .withColumn("_source_endpoint",lit("orders"))
    df.coalesce(1).write.format("csv").option("header",True).mode("append").save("blue-owls-de-assessment/submission/output/bronze/orders.csv")
```
